```python
import re

# Define the function
def find_cancer_types(info):
    # Create a case-insensitive pattern
    pattern = "|".join(re.escape(cancer) for cancer in cancer_types)
    matches = re.findall(pattern, info, flags=re.IGNORECASE)  # Use IGNORECASE flag
    
    # Normalize matches to match the case in `cancer_types`
    normalized_matches = {c.lower(): c for c in cancer_types}  # Create a lookup dictionary
    unique_matches = set(normalized_matches.get(match.lower(), match) for match in matches)
    
    # Return a comma-separated string of unique matches
    return ", ".join(sorted(unique_matches)) if unique_matches else None

# Apply function to DataFrame
df['cancer_type'] = df['Info'].apply(find_cancer_types)

df.head()
```

```python
df = df.dropna()
```
